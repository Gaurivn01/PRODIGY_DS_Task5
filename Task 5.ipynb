```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('RTA Dataset.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking for duplicates/missing values
df.isnull().sum()
```

```python
#Cleaning
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
#cleaned
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Sex_of_driver',data = df)
plt.title('Accident Count based on Gender')
plt.show()
```

```python
road_cond_count = df['Road_surface_type'].value_counts()
weather_cond_count = df['Weather_conditions'].value_counts()
light_cond_counts = df['Light_conditions'].value_counts()

#Road conditions
plt.figure(figsize = (7,7))
sns.barplot(x=road_cond_count.index , y = road_cond_count.values)
plt.title('Distribution based on Road conditions')
plt.xlabel('Road conditions')
plt.ylabel('Number of accidents')
plt.show()

#weather conditions
plt.figure(figsize = (7,7))
sns.barplot(x=weather_cond_count.index , y = weather_cond_count.values)
plt.title('Distribution based on Weather conditions')
plt.xlabel('Weather conditions')
plt.ylabel('Number of accidents')
plt.show()

#Light conditions
plt.figure(figsize = (7,7))
sns.barplot(x=light_cond_counts.index , y = light_cond_counts.values)
plt.title('Distribution based on Light conditions')
plt.xlabel('Light conditions')
plt.ylabel('Number of accidents')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Age_band_of_driver'],bins=20,kde=True,color='skyblue')
plt.title('Distribution of accident by age')
plt.xlabel('age')
plt.ylabel('Number of accidents')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Type_of_collision'],bins=20,kde=True,color='skyblue')
plt.title('Types of collisions')
plt.xlabel('Collisions')
plt.ylabel('Number of accidents')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
contingency_table = pd.crosstab(df['Driving_experience'],df['Accident_severity'])
sns.heatmap(contingency_table,annot=True)
plt.title('Heatmap')
plt.show()
```

```python
#Radar chart - jn type vs collison
junc_coll = df.groupby('Types_of_Junction')['Type_of_collision'].value_counts().unstack().plot(kind='bar',stacked=True)
plt.title('Radar chart')
plt.xlabel('Types of Junctions')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['Accident_severity'] = pd.to_numeric(df['Accident_severity'], errors='coerce')

# Now convert the column to float
df['Accident_severity'] = df['Accident_severity'].astype(float)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt


# Extract latitude and longitude columns
latitude = df['Sex_of_driver']
longitude = df['Accident_severity']

# Create Point geometries
geometry = [Point(xy) for xy in zip(longitude, latitude)]

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Plot the data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
ax = world.plot(figsize=(10, 8), color='gray')
gdf.plot(ax=ax, color='red', markersize=5, alpha=0.5)
plt.title('RTA Dataset - Geospatial Plot')
plt.xlabel('Sex of driver')
plt.ylabel('Severity')
plt.show()
```

```python
sns.pairplot(df,markers='o')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of RTA Dataset')
plt.show()
```

```python
weather_road_surface_df = df[['Weather_conditions', 'Road_surface_type']]

# Encode categorical variables as numerical for correlation analysis
weather_road_surface_encoded = pd.get_dummies(weather_road_surface_df)

# Compute the correlation matrix
corr_matrix = weather_road_surface_encoded.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Weather vs. Road Surface Type')
plt.show()
```
